```python
dataframes = [df1, df2, df3, df4, df5, df6, df7]
```

```python
from matplotlib.animation import FuncAnimation
from IPython.core.display import HTML
```

```python
fig, ax = plt.subplots()
def ppg_func(f_name):
    global dataframes
    ppg = {}
    colors = ["b", "g", "r", "c", "m", "y", "k"]
    data = dataframes[f_name]
    for row in range(len(data)):
        index=int(data.iloc[row]["Pk"])
        if data.iloc[row]["G"] != 0:
            ppg[index] = (float(data.iloc[row]["PTS"])/float(data.iloc[row]["G"]))
        else:
            ppg[index] = 0
    ax.clear()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.scatter(range(len(ppg)), list(ppg.values()), color=colors[f_name])
    ax.set_xlabel("Draft Pick")
    ax.set_xlim(-2, 62)
    ax.set_ylim(-2, 27)
    ax.set_ylabel("Points per Game") #this is over the entire set of data from 2013-2019
    ax.set_title("201"+str(f_name+3)+" Draft Class", color=colors[f_name])
    
anim = FuncAnimation(fig, ppg_func, frames=len(dataframes), interval=1500)

html = anim.to_html5_video()
with open("ppg_draft.html", "w") as f:
    f.write(html)
plt.close(fig)
HTML(html)
```

```python
fig, ax = plt.subplots()
def fieldg_func(f_name):
    global dataframes
    fgp = {}
    colors = ["b", "g", "r", "c", "m", "y", "k"]
    data = dataframes[f_name]
    for row in range(len(data)):
        index=int(data.iloc[row]["Pk"])
        if data.iloc[row]["G"] != 0:
            fgp[index] = (float(data.iloc[row]["FG%"]))
        else:
            pass
    ax.clear()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_ylim(-0.05, .8)
    ax.set_xlim(-2, 62)
    ax.scatter(range(len(fgp)), list(fgp.values()), color=colors[f_name])
    ax.set_xlabel("Draft Pick")
    ax.set_ylabel("Field Goal Percent") #this is over the entire set of data from 2013-2019
    ax.set_title("201"+str(f_name+3)+" Draft Class", color=colors[f_name])
    
anim = FuncAnimation(fig, fieldg_func, frames=len(dataframes), interval=1500)

html = anim.to_html5_video()
with open("fg_draft.html", "w") as f:
    f.write(html)
plt.close(fig)
HTML(html)
```
